### DATA ANALYSIS PROJECT : RESTAURENT TIP ANALYSIS 

### OBJECTIVES
- Do male tip more than female?
- Does day effect tip?
- Does bigger group gives larger tip?
- Does time(lunch/Dinner) effect tip?

```python
# importing libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# gathering information
df.info()
```

```python
# Dataset numerical values
df.describe()
```

```python
# total null values in this dataset
print("null value:",df.isnull().sum().sum())
```

```python
# columns in dataset
df.columns
```

### TOTAL_BILL DISTRIBUTION

```python
plt.figure(figsize=(6,3))
sns.histplot(df['total_bill'],kde=True)
plt.show()
```

### Total_bill vs Tip

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.show()
```

#### Tips by gender

```python
plt.figure(figsize=(6,3))
sns.boxplot(x='sex',y='tip',data=df)
plt.title("Tips By Gender")
plt.xlabel("Gender")
plt.ylabel("Tips")
plt.show()
```

### TIPS BY DAY

```python
plt.figure(figsize=(6,3))
sns.boxplot(x='day',y='tip',data=df,palette='coolwarm')
plt.title("Tips By Day")
plt.xlabel("days")
plt.ylabel("Tips")
plt.show()
```

#### TIP BY TIME

```python
plt.figure(figsize=(6,3))
sns.boxplot(x='time',y='tip',data=df,palette='viridis')
plt.title("Tips By Time")
plt.xlabel("Time")
plt.ylabel("Tips")
plt.show()
```

```python
### Pair Plot Analysis
```

```python
plt.figure(figsize=(6,3))
sns.pairplot(df)
plt.show()
```

#### CORRELATION HEATMAP

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(6,3))
sns.heatmap(corr , annot=True , cmap='coolwarm')
plt.show()
```
